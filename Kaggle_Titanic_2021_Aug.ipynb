## **Imputing Missing Values**
---


Imputation needed for :
* Cabin -- strip out prefixed letter and group nulls   
* Embarkation when missing. Impute with mode
* Age when missing. Median Age by Salutation
* Fares when 0. Median of Salutation and Ticket Class




```python
# Checking pattern of Cabin Name. Prefixed letter seems important.
cabins = df_train['Cabin Name'].tolist() 
cabins[:20]
```

```python
#Prefixed letter retained
df_train['Cabin Letter'] = np.where(df_train['Cabin Name'] != False,df_train['Cabin Name'].str[:1],df_train['Cabin Name'])
print(df_train['Cabin Letter'])
```

```python
#Group together missing into "NA"
df_train['Cabin Letter'].fillna('NA', inplace = True)
print(df_train['Cabin Letter'])
```

```python
# Dropping Cabin Name since Cabin Letter is processed
df_train = df_train.drop(columns=['Cabin Name'])
```

```python
# Just checking that Cabin Name was dropped
list(df_train.columns)
```

```python
#Impute most frequent port for missing
df_train['Port of Embarkation'].fillna(df_train['Port of Embarkation'].mode()[0], inplace = True)
```

```python
# Ensuring Port imputation was successful
df_train['Port of Embarkation'].isnull().sum()
```

```python
#Stripping out salutation from name
df_train['Salutation'] = df_train['Name'].apply(lambda name: name.split(',')[1].split('.')[0].strip())
```

```python
# Checking that Salutation was successfully created
print(pd.unique(df_train['Salutation']))
```

```python
# Calculate the average fare by Salutation and Ticket Class
fare = df_train.groupby(['Salutation','Ticket Class'])[['Passenger Fare']].mean()
```

```python
# Ensuring that the processed fare is greater than the original
print('Original :',df_train['Passenger Fare'].mean())
print('Processed :',fare['Passenger Fare'].mean())
```

```python
# Merge that calculation onto the original dataframe
df_train = df_train.merge(fare.reset_index(),on=['Salutation','Ticket Class'],suffixes=('','_Right'))
```

```python
# Having a look at those w/ 0 Passenger Fare vs Passenger Fare_Right average. Only one Jonkheer Salutation in population, so no imputation possible.
df_train.loc[df_train['Passenger Fare']==0] 
```

```python
# Replacing the 0 fare's with the average fare
df_train.loc[df_train['Passenger Fare']==0, 'Passenger Fare'] = df_train.loc[df_train['Passenger Fare']==0, 'Passenger Fare_Right']
```

```python
# Checking that only the one record remains at 0
df_train.loc[df_train['Passenger Fare']==0]
```

```python
# Min is still 0 because of the Jonkheer
df_train['Passenger Fare'].describe()
```

```python
# Checking things out
df_train.describe()
```

```python
# Dropping the average column
df_train = df_train.drop(columns=['Passenger Fare_Right'])
```

```python
# Create a column for count of family members
df_train['Family'] = df_train['Sibling Count'] + df_train['Parent Count']
```

```python
# Giving things a look. Only Age remains as needing imputation!
df_train.count()
```

```python
# Calculating the average age by Salutation
age = df_train.groupby('Salutation')[['Age']].mean()
```

```python
# Affixing that calculation to the dataframe
df_train = df_train.merge(age.reset_index(),on=['Salutation'],suffixes=('','_Right'))
```

```python
# Checking things out
df_train[df_train['Age'].isnull()]
```

```python
# Just giving things a look
print('Original :',df_train['Age'].mean())
print('Processed :',df_train['Age_Right'].mean())
```

```python
# Replacing nulls with 0 in order to convernt to interger
df_train['Age'] = df_train['Age'].fillna(0)
```

```python
# No longer any nulls
df_train[df_train['Age'].isnull()]
```

```python
# Converting to integer
df_train['Age'] = df_train['Age'].astype(int)
df_train.info()
```

```python
# Replace 0's with average
df_train.loc[df_train['Age']==0, 'Age'] = df_train.loc[df_train['Age']==0, 'Age_Right']
```

```python
# No longer any 0's
df_train.loc[df_train['Age']==0]
```

```python
# Giving it a look
df_train.describe()
```

```python
# Drop extra column and give it a final check
df_train = df_train.drop(columns=['Age_Right'])
df_train.describe()
```

##Trying out h2o

```python
import h2o
from h2o.automl import H2OAutoML
h2o.init()
```

```python
# Import a sample binary outcome train/test set into H2O
train = h2o.H2OFrame(df_train_curated_80)
test = h2o.H2OFrame(df_train_curated_20)

# Identify the response and set of predictors
y = 'Survived'
x = list(train.columns)  #if x is defined as all columns except the response, then x is not required
x.remove(y)
x = train.columns 
x.remove(y) 

# For binary classification, response should be a factor
train[y] = train[y].asfactor()
test[y] = test[y].asfactor()
# Run AutoML for 30 seconds
aml = H2OAutoML(max_runtime_secs=120, seed=17)
aml.train(x=x,y=y, training_frame=train)

# Print Leaderboard (ranked by xval metrics)
aml.leaderboard
# (Optional) Evaluate performance on a test set
perf = aml.leader.model_performance(test)
perf.auc()
```

```python
# Set up an H2OAutoML object
aml = H2OAutoML(max_runtime_secs=30)
# Launch an AutoML run
aml.train(y=y, training_frame=train)
# Get the best model in the AutoML Leaderboard
aml.leader

# Get AutoML object by `project_name`
get_aml = h2o.automl.get_automl(aml.project_name)
# Get the best model in the AutoML Leaderboard
get_aml.leader
```

```python
lb = h2o.automl.get_leaderboard(aml, extra_columns = 'ALL')
lb
```

```python
holdout = h2o.H2OFrame(df_test_curated)
# Set up an H2OAutoML object
aml = H2OAutoML(max_runtime_secs=30)
# Launch an H2OAutoML run
aml.train(y=y, training_frame=train)
# Predict with top model from AutoML Leaderboard on a H2OFrame called 'test'
aml.predict(holdout)

# Get AutoML object by `project_name`
get_aml = h2o.automl.get_automl(aml.project_name)
# Predict with top model from AutoML Leaderboard on a H2OFrame called 'test'
get_aml.predict(holdout)
```

```python
submission_h2o = get_aml.predict(holdout)
```

```python
submission_h2o_df = h2o.as_list(submission_h2o)
submission_h2o_concat = pd.concat([submission_h2o_df, df_test_curated], axis=1)
submission_h2o_concat
```
